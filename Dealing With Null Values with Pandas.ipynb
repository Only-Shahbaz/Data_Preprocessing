# Import Libraries

````python
import numpy as np
import pandas as pd
````

 Python list <b>headers</b> containing name of headers


````python
# These are the headers which you need to link to your dataset.
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
````

# Import Datasets

Use the Pandas method <b>read_csv()</b> to load the data from the web address. Set the parameter  "names" equal to the Python list "headers".


````python
df = pd.read_csv('car_dataset.data', names = headers)
````

 Use the method <b>head()</b> to display the first five rows of the dataframe.


````python
# To see what the data set looks like, we'll use the head() method.
df.head(15)
````

<h1 id="identify_handle_missing_values">Identify and handle missing values</h1>

<h2 id="identify_missing_values">Identify missing values</h2>
<h3>Convert "?" to NaN</h3>
In the car dataset, missing data comes with the question mark "?".
We replace "?" with NaN (Not a Number), which is Python's default missing value marker, for reasons of computational speed and convenience. Here we use the function:
 <pre>.replace(A, B, inplace = True) </pre>
to replace A by B


````python
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df
````

Identify_missing_values

<h2>Evaluating for Missing Data</h2>

The missing values are converted to default. We use the following functions to identify these missing values. There are two methods to detect missing data:

<ol>
    <li><b>.isnull()</b></li>
    <li><b>.notnull()</b></li>
</ol>
The output is a boolean value indicating whether the value that is passed into the argument is in fact missing data.


````python
df.isnull()
````

"True" stands for missing value, while "False" stands for not missing value.


<h2>Count missing values in each column</h2>
<p>
Using a for loop in Python, we can quickly figure out the number of missing values in each column. As mentioned above, "True" represents a missing value, "False"  means the value is present in the dataset.  In the body of the for loop the method  ".value_counts()"  counts the number of "True" values.
</p>


````python
missing_data = df.isnull()
for columns in headers:
  print(columns)
  print(missing_data[columns].value_counts())
  print(" ")
````

Based on the summary above, each column has 205 rows of data, seven columns containing missing data:

<ol>
    <li>"normalized-losses": 41 missing data</li>
    <li>"num-of-doors": 2 missing data</li>
    <li>"bore": 4 missing data</li>
    <li>"stroke" : 4 missing data</li>
    <li>"horsepower": 2 missing data</li>
    <li>"peak-rpm": 2 missing data</li>
    <li>"price": 4 missing data</li>
</ol>


<h2 id="deal_missing_values">Deal with missing data</h2>
<br>
<h3>How to deal with missing data?</h3>

<ol>
    <li>drop data<br>
        a. drop the whole row<br>
        b. drop the whole column
    </li>
    <li>replace data<br>
        a. replace it by mean<br>
        b. replace it by frequency<br>
    </li>
</ol>


Whole columns should be dropped only if most entries in the column are empty. In our dataset, none of the columns are empty enough to drop entirely.
We have some freedom in choosing which method to replace data; however, some methods may seem more reasonable than others. We will apply each method to many different columns:

<b>Replace by mean:</b>

<ul>
    <li>"normalized-losses": 41 missing data, replace them with mean</li>
    <li>"stroke": 4 missing data, replace them with mean</li>
    <li>"bore": 4 missing data, replace them with mean</li>
    <li>"horsepower": 2 missing data, replace them with mean</li>
    <li>"peak-rpm": 2 missing data, replace them with mean</li>
</ul>

<b>Replace by frequency:</b>

<ul>
    <li>"num-of-doors": 2 missing data, replace them with "four".
        <ul>
            <li>Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur</li>
        </ul>
    </li>
</ul>

<b>Drop the whole row:</b>

<ul>
    <li>"price": 4 missing data, simply delete the whole row
        <ul>
            <li>Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful to us</li>
        </ul>
    </li>
</ul>


<h3>Calculate the average of the normalized-losses column </h3>


````python
mean_value = df["normalized-losses"].astype("float").mean()
mean_value
````

<h3>Replace "NaN" by mean value in "normalized-losses" column</h3>


````python
df["normalized-losses"].replace(np.nan, mean_value, inplace = True)
df
````

<h3>Calculate the mean value for 'bore' column</h3>


````python
mean_value = df["bore"].astype("float").mean()
mean_value
````

<h3>Replace NaN by mean value for bore column</h3>


````python
df["bore"].replace(np.nan, mean_value, inplace = True)
df
````

<h3>Calculate the mean value for 'stroke' column</h3>

````python
# Write your code below and press Shift+Enter to execute
mean_value = df['stroke'].astype(float).mean(axis=0)
mean_value
````

<h3>Replace NaN by mean value for stroke column</h3>

````python
df['stroke'].replace(np.nan, mean_value, inplace = True)
df
````

<h4>Calculate the mean value for the  'horsepower' column:</h4>


````python
mean_value = df['horsepower'].astype(float).mean(axis=0)
mean_value
````

<h4>Replace "NaN" by mean value:</h4>


````python
df['horsepower'].replace(np.nan, mean_value, inplace = True)
df
````

<h4>Calculate the mean value for 'peak-rpm' column:</h4>


````python
mean_val = df['peak-rpm'].astype("float").mean(axis=0)
mean_val
````

<h4>Replace NaN by mean value:</h4>


````python
df["peak-rpm"].replace(np.nan, mean_val, inplace = True)
````


<ul>
<li>Now we have done for replacing with mean.</li>
<li>Its time to replace null values with most frequent values </li>
<li>We will do this for num-of-doors column</li>
</ul>

<h4>Find the most frequent value for num of doors</h4>

````python
most_used = df['num-of-doors'].value_counts()
most_used
````

<h4>We can see that four doors are the most common type. We can also use the ".idxmax()" method to calculate for us the most common type automatically:</h4>


````python
most_used = df['num-of-doors'].value_counts().idxmax()
most_used
````

<h4>The replacement procedure is very similar to what we have seen previously</h4>


````python
#replace the missing 'num-of-doors' values by the most frequent
df['num-of-doors'].replace(np.nan, most_used, inplace = True)
````

## Finally, let's drop all rows that do not have price data:


````python

# simply drop whole row with NaN in "price" column
df.dropna(subset = "price", inplace = True, axis= 0)

# reset index, because we droped two rows
df.reset_index(drop = True, inplace = True)
````

Here ***subset*** parameter enable us to specify the column name in which you want to drop the nan values ***inplace = True*** means you want to make these chances in the current dataframe and ***axis = 0*** parameter specify that you want drop the entire row.  

<h2> Lets Check It</h2>

````python
missing_data = df.isnull()
for columns in headers:
  print(columns)
  print(missing_data[columns].value_counts())
  print(" ")
````

We have drop the 4 rows in the price column with missing values, now the data has no null values in it and is ready for further analysis.
